#Exploratory Data Analysis using Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("Titanic.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)

df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

df["Cabin"].fillna("Unknown", inplace=True)

df.isnull().sum()
```

**Q1. Create a countplot to show the number of passengers by gender.**

```python
plt.figure(figsize=(7,5))

sns.countplot(x="Sex", data=df)

plt.title("Number of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.show()
```

**Interpretation**
* The chart shows the number of male and female passengers.
* Male passengers are more than female passengers.

**Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.**

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x="Pclass",
    data=df,
    palette=["skyblue","orange","green"]
)

plt.title("Number of Passengers by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

**Interpretation**
* Third-class passengers are the largest group.
* First-class passengers are fewer than third-class passengers.

**Q3. Create a histogram to show the distribution of passenger age.**

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=30,
    kde=True,
    color="purple"
)

plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

**Interpretation**
* Most passengers were between 20 and 40 years old.
* The age distribution is slightly right-skewed.

**Q4. Create a boxplot to detect outliers in the fare column.**

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=df["Fare"],
    color="gold"
)

plt.title("Boxplot of Fare")

plt.show()
```

**Interpretation**
* Several outliers are visible.
* Some passengers paid much higher fares than the majority.

**Q5. Create a scatterplot to**


**a. show the relationship between age and fare.**

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="Age",
    y="Fare",
    data=df,
    alpha=0.7
)

plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
correlation=df["Age"].corr(df["Fare"])

print("Correlation =",round(correlation,3))
```

**Interpretation**
* The scatter plot shows a weak relationship between Age and Fare.
* Passenger age does not strongly affect ticket fare.

**b. analyze the relationship between family size (SibSp + Parch) and fare.**

```python
df["Family_Size"]=df["SibSp"]+df["Parch"]
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="Family_Size",
    y="Fare",
    data=df,
    color="green",
    alpha=0.7
)

plt.title("Family Size vs Fare")
plt.xlabel("Family Size")
plt.ylabel("Fare")

plt.show()
```

**Interpretation**
* Family size has only a weak relationship with fare.
* Some larger families paid higher fares, but the relationship is not very strong.

**Q6. Create a countplot to show survival count by gender.**

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Sex",
    hue="Survived",
    data=df
)

plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Passengers")

plt.legend(title="Survived")

plt.show()
```

**Interpretation**
* Female passengers had a much higher survival rate than male passengers.
* Male passengers had the highest number of deaths.

**Q7. Create a bar chart to show survival by passenger class.**

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Pclass",
    y="Survived",
    data=df
)

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Survival Rate")

plt.show()
```

**Interpretation**
* First-class passengers had the highest survival rate.
* Third-class passengers had the lowest survival rate.

**Q8. Create a heatmap to show correlation between numerical variables.**

```python
plt.figure(figsize=(10,7))

numeric=df.select_dtypes(include=np.number)

sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()
```

**Interpretation**
* Fare has a positive correlation with survival.
* Passenger class has a negative correlation with survival.
* Larger passenger class numbers (lower travel class) generally have lower survival rates.

**Q9. Identify which passenger class paid the highest fare on average.**

```python
average_fare=df.groupby("Pclass")["Fare"].mean()

print(average_fare)
```

```python
plt.figure(figsize=(7,5))

average_fare.plot(
    kind="bar",
    color=["blue","green","red"]
)

plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")

plt.show()
```

```python
highest_class=average_fare.idxmax()

print("Passenger Class Paying Highest Average Fare:",highest_class)
```

**Interpretation**
* Passenger Class 1 paid the highest average fare.
* First-class tickets were significantly more expensive than second- and third-class tickets.
